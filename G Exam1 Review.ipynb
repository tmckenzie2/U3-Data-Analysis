## Exam Coverage
The exam covers everything we have covered so far in the course. Here is a general outline of the topics we have covered so far (note that this list is not comprehensive of everything we have covered so far, but a general outline):

### U0 Introduction
1. What is data mining?
1. Why use Python programming for data mining?
    1. Why Anaconda Python Distribution
    1. Working with Anaconda/Python related software
    1. Interactive vs. scripting Python
1. Terms related to datasets
    1. Tabular data, CSV files
    1. Rows, columns, instances, attributes, keys (primary and foreign), etc.
    1. Performing joins
1. Common operating systems commands
1. Running Python from command line


### U1 Python
1. Python basics
    1. Variables, types, comments, conditionals, loops, functions/methods, file IO, strings, etc.
    1. Significance of `__name__`
    1. Working with modules (standard libraries and programmer-defined)
        1. math, copy, operator, etc.
        1. Why do we create a utils module?
1. Working with Python data structures 
    1. Lists, tuples, dictionaries, sets, etc.
    1. Common algorithms operating on 1-D, 2-D, and N-D lists
1. Pass by object reference
    1. Alias vs. shallow copy vs. deep copy


### U2 Git/Github
1. What is a version control system and why do we use it?
1. What is Git?
1. What is Github?
1. What does it mean to clone, commit, push, and pull?
1. Using Git/Github from command line


### U3 Data Preprocessing
1. What is data preprocessing?
    1. What are the steps? Exploration, reduction, integration, cleaning, transformation...
1. Data types/measurement scales
    1. Nominal, ordinal, interval, ratio
    1. Categorical vs. continuous
    1. Noisy vs. invalid
    1. Labeled vs. unlabeled
    1. Supervised vs. unsupervised learning
1. How do we deal with missing data?
1. Summary statistics
    1. Midrange, mean, median, mode, variance, standard deviation, quantiles, etc.
1. Computing frequencies
1. Group by and its uses
1. Discretization approaches
    1. Meaningful ranges, equal width, equal frequency
    1. Computing equal width cutoff values and bin frequencies
1. Data visualization
    1. Different types of charts and what they are used for
    1. Basic plotting with matplotlib
        1. Note: you don't need to know API details, just big picture concepts
1. Linear regression
    1. Basic formula, how to calculate, how to use
    1. Correlation, covariance
1. Jupyter Notebook basics
    1. Why Jupyter Notebook?
    1. Why data science storytelling?
    1. Markdown basics
    1. Latex basics

The exam questions will require you to solve problems in a variety of formats: defining key terms/concepts, applying algorithms to example datasets, writing code, determining the output of code, etc.

* Review reading quizzes, warm-up tasks, programming assignments, problems solved in class, and code derived in class
    * These may well be your best resources
    * An excellent learning activity would be to re-solve these exercises
* Lesson notes
* Re-read the pages assigned for the reading quizzes
    * Solve the problems included at the end of the chapters
